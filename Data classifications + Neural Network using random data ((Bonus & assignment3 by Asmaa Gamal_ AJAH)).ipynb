```python
#--------------------------------------------------------------   input & output initialization  ------------------------------------------------------------------------------------------------
#first splitting of my data to initialize x as an input and y as an output
import pandas as pd

split = pd.DataFrame(dataset)
  
split.head()
x_imbalanced = split.iloc[:,:-1]  #x= all columns except the last column
y_imbalanced = split.iloc[:,-1 ]  #y= the last column only


print(x_imbalanced)
print(y_imbalanced)

```

```python
#------------------------------------------------------------------------  Data Balancing ----------------------------------------------------------------------------------------------------------------


from collections import Counter
from imblearn.under_sampling import ClusterCentroids 
cc = ClusterCentroids()
x, y = cc.fit_resample(x_imbalanced, y_imbalanced)
print('Balanced dataset shape %s' % Counter(y))

'''
#saving to avoid time delay to do resampliing each time we run our code:
balanced_data = np.concatenate((x, y), axis = 0)

data.savetxt('Balanced dataset.csv', balanced_data, delimiter=",")
'''

```

```python
#---------------------------------------------------------------------training & testing sets-------------------------------------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y ,random_state=0, test_size=0.3)
print(x_train.shape) #70% training
print(x_test.shape)  #30% testing
```

```python
#------------------------------------------------------------------------classification-------------------------------------------------------------------------
```

```python
#----------------------------------------------------------------First classifier: Decision Tree------------------------------------------------------
#--------------------------------------------------------------No parameters need to be tuned here -------------------------------------------------------

from sklearn          import tree


#classifier
DTree           = tree.DecisionTreeClassifier()

#train
DTree_trained   = DTree.fit(x_train,y_train)


#test
print(DTree_trained.predict(x_test))
print(y_test)
```

```python
#plotting the full tree
tree.plot_tree(DTree_trained ) 
```

```python
#printing the classification report 
from sklearn.metrics import classification_report , confusion_matrix

print('the classification report of the Decision Tree classifier without tunning parameters' )
print(classification_report(DTree_trained.predict(x_test) ,y_test ))
print( 'Confusion Matrix Array =' , confusion_matrix(y_test, DTree_trained.predict(x_test)) )


```

```python
#-------------------------------------------------------------------Second classifier: AdaBoost ---------------------------------------------------------------------
#-----------------------------------------------------------Tunning parameters using cross validation-------------------------------------------------------
from sklearn.model_selection import cross_val_score
from sklearn.ensemble        import AdaBoostClassifier



# empty list that will hold cv scores
crossv_scores = []

#best n 
n_estimators = [i for i in range(10,200,10)]


for n in n_estimators:
    adaBoost          = AdaBoostClassifier(n_estimators=n)
    scores            = cross_val_score(adaBoost, x, y, cv=10, scoring='accuracy')
    crossv_scores.append(scores.mean())




#misclassification error
misclass_error = [1-i for i in crossv_scores]

#optimal K
optimal_n_index = misclass_error.index(min(misclass_error))
optimal_n = n_estimators[optimal_n_index]
print('optimal n estimators =',optimal_n)
print('misclassification error in this case=',min(misclass_error))

#graphical view
import matplotlib.pyplot as plt
# plotting the error vs k
plt.plot(n_estimators,misclass_error)
plt.xlabel('Number of n_estimators')
plt.ylabel('Misclassification Error ')
plt.show()
```

```python
#-------------------------------------------------------Classification results after Tunning parameters-------------------------------------------------------
#-------------------------------------------------------------------Second classifier: AdaBoost ---------------------------------------------------------------------
from sklearn.ensemble import AdaBoostClassifier


#classifier
AdaBoost          = AdaBoostClassifier(n_estimators=optimal_n)

#train
AdaBoost_trained   = AdaBoost.fit(x_train,y_train)


#test
print(AdaBoost_trained.predict(x_test))
print(y_test)
```

```python
#printing the classification report 
from sklearn.metrics import classification_report ,confusion_matrix
print('the classification report of the AdaBoost classifier after tunning parameters' )
print(classification_report(AdaBoost_trained.predict(x_test) ,y_test ))
print( 'Confusion Matrix Array =' , confusion_matrix(y_test, AdaBoost_trained.predict(x_test)) )
```

```python
#-------------------------------------------------------------------third classifier: K-NN---------------------------------------------------------------------
#-----------------------------------------------------------Tunning parameters using cross validation-------------------------------------------------------

from sklearn.model_selection import cross_val_score
from sklearn.neighbors       import KNeighborsClassifier



# empty list that will hold cv scores
crossv_scores = []

#best K will be an odd num 
neighbors = [i for i in range(1,50) if i % 2 != 0]


for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=10, scoring='accuracy')
    crossv_scores.append(scores.mean())




#misclassification error
misclass_error = [1-i for i in crossv_scores]

#optimal K
optimal_k_index = misclass_error.index(min(misclass_error))
optimal_k = neighbors[optimal_k_index]
print('optimal_k =',optimal_k)

#graphical view
import matplotlib.pyplot as plt
# plotting the error vs k
plt.plot(neighbors,misclass_error)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error ')
plt.show()
```

```python
#-------------------------------------------------------Classification results after Tunning parameters-------------------------------------------------------
#-------------------------------------------------------------------third classifier: K-NN---------------------------------------------------------------------
from sklearn.neighbors import KNeighborsClassifier


#classifier
KNN           = KNeighborsClassifier(n_neighbors=optimal_k)

#train
KNN_trained   = KNN.fit(x_train,y_train)


#test
print(KNN_trained.predict(x_test))
print(y_test)
```

```python
#printing the classification report 
from sklearn.metrics import classification_report, confusion_matrix
print('the classification report of the K-NN classifier with tunning the parameters' )
print(classification_report(KNN_trained.predict(x_test) ,y_test ))
print( 'Confusion Matrix Array =' , confusion_matrix(y_test, KNN_trained.predict(x_test)) )
```

```python
#-------------------------------------------------------------------fourth classifier: Random forest---------------------------------------------------------------------
#-----------------------------------------------------------Tunning parameters using cross validation--------------------------------------------------------------------
from sklearn.model_selection import cross_val_score
from sklearn.ensemble        import RandomForestClassifier

# empty list that will hold cv scores
crossv_scores = []

#best n 
nForest_estimators = [i for i in range(10,200,10)]


for n in nForest_estimators:
    random_f          = RandomForestClassifier(n_estimators=n)            

    scores            = cross_val_score(random_f, x, y, cv=10, scoring='accuracy')
    crossv_scores.append(scores.mean())




#misclassification error
misclass_error = [1-i for i in crossv_scores]

#optimal n
optimal_n_index = misclass_error.index(min(misclass_error))
optimal_n_forest = nForest_estimators[optimal_n_index]
print('optimal n estimators of the Random forest classifier=',optimal_n_forest)
print('misclassification error in this case=',min(misclass_error))





#graphical view
import matplotlib.pyplot as plt

# plotting the error vs k
plt.plot(nForest_estimators,misclass_error)
plt.xlabel('Number of n_estimators')
plt.ylabel('Misclassification Error ')
plt.show()
```

```python
#-------------------------------------------------------Classification results after Tunning parameters-------------------------------------------------------
#-------------------------------------------------------------------fourth classifier: Random forest---------------------------------------------------------------------

from sklearn.ensemble import RandomForestClassifier


#classifier
RandomF           = RandomForestClassifier(n_estimators=optimal_n_forest)

#train
RandomF_trained   = RandomF.fit(x_train,y_train)


#test
print(RandomF_trained.predict(x_test))
print(y_test)
```

```python
#printing the classification report 
from sklearn.metrics import classification_report,confusion_matrix
print('the classification report of the random Forest classifier without tunning parameters' )
print(classification_report(RandomF_trained.predict(x_test) ,y_test ))
print( 'Confusion Matrix Array =' , confusion_matrix(y_test, RandomF_trained.predict(x_test)) )

```

```python
#------------------------------------------------------------------------------Neural Network------------------------------------------------
#________________________Step 1

import torch 
import torch.nn as nn


#________________________Step 2
# Defining input size, hidden layer size, output size and batch size respectively
features_columns=10
n_hidden=2
n_out=1
train_examples_raws = 10



#________________________Step 3

# Create dummy data:
x_copy = torch.randn(train_examples_raws, features_columns)
y_copy = torch.tensor([[1.0], [0.0], [0.0], [1.0], [1.0], [1.0], [0.0], [0.0], [1.0], [1.0]])
#print(type(y_copy))

'''#transform Data into torch 
from numpy import asarray
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder

y_copy,x_copy= y,x

# transform y from g or h to 0 or 1
y_copy= y_copy.reshape(2, 6688)
encoded=OrdinalEncoder()                                                       #this is to create an obiect from this class  # i can also use:  OneHotEncoder(drop='second',sparse=False)   #drop='second' so insteed of having two num 0,0 or 0,1  as my output, i will have one num 0 or 1

y_copy= encoded.fit_transform(y_copy)
#return to right shape
y_copy=y_copy.reshape(13376,1)
y_copy=torch.from_numpy(y_copy)
x_copy=torch.from_numpy(x_copy)
#splitting
'''
#---------------------------------------------------------------------training & testing sets-------------------------------------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split

x_torchtrain, x_torchtest, y_torchtrain, y_torchtest = train_test_split(x_copy,y_copy ,random_state=0, test_size=0.3)
print('x_torch_train.shape=',x_torchtrain.shape) #70% training
print('x_torch_test.shape=' ,x_torchtest.shape )  #30% testing






#________________________Step 4
# Create a model
model = nn.Sequential(   nn.Linear(features_columns, n_hidden)    ,     nn.ReLU()    ,      nn.Linear(n_hidden, n_out)    ,        nn.Sigmoid()    )



#________________________Step 5
#Construct the loss function
criterion = torch.nn.MSELoss()
# Construct the optimizer (Stochastic Gradient Descent in this case)
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)







#________________________Step 6
# Gradient Descent
all_losses=[]
tot_iterations=[i for i in range(50)]

for i in range(len(tot_iterations)):
   # Forward pass: Compute predicted y by passing x to the model
   y_pred = model(x_torchtrain)
   ypred_test = model(x_torchtest)

   # Compute and print loss
   loss      = criterion(y_pred, y_torchtrain)
   loss_test = criterion(ypred_test, y_torchtest)
   all_losses.append(loss)
   
   print('iteration: ', i,' loss: ', loss.item())

   # Zero gradients, perform a backward pass, and update the weights.
   optimizer.zero_grad()

   # perform a backward pass (backpropagation)
   loss.backward()

   # Update the parameters
   optimizer.step()


#graphical view
import matplotlib.pyplot as plt
# plotting the error vs iteration 
plt.plot(tot_iterations,all_losses)
plt.xlabel('Number of iterations in this neural network model')
plt.ylabel('Misclassification Error ')
plt.show()

   

#------------------------------------------------------------------------Neural network classification report-------------------------------------------------
print(y_torchtest)
print(ypred_test)


import torch

y_torchtest = torch.tensor(y_torchtest,requires_grad=True)
ypred_test  = torch.tensor(ypred_test ,requires_grad=True)

y_torchtest = y_torchtest.detach().numpy()
ypred_test = ypred_test.detach().numpy()
 

y_torchtest=y_torchtest.astype(float)
ypred_test=y_torchtest.astype(float)

print('pred y of my test=\n',ypred_test)
print('real y of my test Group=\n',y_torchtest)



from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_torchtest, ypred_test))
print( 'Confusion Matrix Array =' , confusion_matrix(y_torchtest, ypred_test ))











```
